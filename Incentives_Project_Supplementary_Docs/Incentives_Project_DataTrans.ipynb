```python
# Combine southwest san gabriel, upper san gabriel, and east san gabriel into san gabriel

# Total Population
crash_data19_t['san gabriel'].loc[['t_pop']] = (
    crash_data19_t['southwest san gabriel'].loc[['t_pop']]
    + crash_data19_t['upper san gabriel'].loc[['t_pop']]
    + crash_data19_t['east san gabriel'].loc[['t_pop']]
        )

# Median Age
crash_data19_t['san gabriel'].loc[['med_age']] = statistics.median([
(int(crash_data19_t['southwest san gabriel'].loc[['med_age']])),
(int(crash_data19_t['upper san gabriel'].loc[['med_age']])),
(int(crash_data19_t['east san gabriel'].loc[['med_age']]))])
     

# Only High school graduation (percent of pop %)includes weights
crash_data19_t['san gabriel'].loc[['hs_g']] = np.round((
((int(crash_data19_t['southwest san gabriel'].loc[['t_pop']])
  /int(crash_data19_t['san gabriel'].loc[['t_pop']]))          #weight
    * crash_data19_t['southwest san gabriel'].loc[['hs_g']])  # value
    + ((int(crash_data19_t['upper san gabriel'].loc[['t_pop']])
     /int(crash_data19_t['san gabriel'].loc[['t_pop']]))          #weight
    * int(crash_data19_t['upper san gabriel'].loc[['hs_g']]))     # value
    + ((int(crash_data19_t['east san gabriel'].loc[['t_pop']])
     /int(crash_data19_t['san gabriel'].loc[['t_pop']]))          #weight
    * int(crash_data19_t['east san gabriel'].loc[['hs_g']]))     # value
        ), decimals =2)

# Less than High school graduation (percent of pop %) includes weights
crash_data19_t['san gabriel'].loc[['lhs_g']] = np.round((
((int(crash_data19_t['southwest san gabriel'].loc[['t_pop']])
  /int(crash_data19_t['san gabriel'].loc[['t_pop']]))          #weight
    * crash_data19_t['southwest san gabriel'].loc[['lhs_g']])  # value
    + ((int(crash_data19_t['upper san gabriel'].loc[['t_pop']])
     /int(crash_data19_t['san gabriel'].loc[['t_pop']]))          #weight
    * int(crash_data19_t['upper san gabriel'].loc[['lhs_g']]))     # value
    + ((int(crash_data19_t['east san gabriel'].loc[['t_pop']])
     /int(crash_data19_t['san gabriel'].loc[['t_pop']]))          #weight
    * int(crash_data19_t['east san gabriel'].loc[['lhs_g']]))     # value
        ), decimals =2)

 
# Population Over 15
crash_data19_t['san gabriel'].loc[['pop_15']] = (
    crash_data19_t['southwest san gabriel'].loc[['pop_15']]
    + crash_data19_t['upper san gabriel'].loc[['pop_15']]
    + crash_data19_t['east san gabriel'].loc[['pop_15']]
        )
 
# Median Income
crash_data19_t['san gabriel'].loc[['med_inc']] = statistics.median([
(int(crash_data19_t['southwest san gabriel'].loc[['med_inc']])),
(int(crash_data19_t['upper san gabriel'].loc[['med_inc']])),
(int(crash_data19_t['east san gabriel'].loc[['med_inc']]))])
 
 
 
# Population Used to Determine Poverty
crash_data19_t['san gabriel'].loc[['pop_det_pov']] = (
    crash_data19_t['southwest san gabriel'].loc[['pop_det_pov']]
    + crash_data19_t['upper san gabriel'].loc[['pop_det_pov']]
    + crash_data19_t['east san gabriel'].loc[['pop_det_pov']]
        )

# Percent of Population under Poverty Line (%) including weights
crash_data19_t['san gabriel'].loc[['pov_level']] = np.round((
((int(crash_data19_t['southwest san gabriel'].loc[['pop_det_pov']])
  /int(crash_data19_t['san gabriel'].loc[['pop_det_pov']]))          #weight
    * crash_data19_t['southwest san gabriel'].loc[['pov_level']])  # value
    + ((int(crash_data19_t['upper san gabriel'].loc[['pop_det_pov']])
     /int(crash_data19_t['san gabriel'].loc[['pop_det_pov']]))          #weight
    * int(crash_data19_t['upper san gabriel'].loc[['pov_level']]))     # value
    + ((int(crash_data19_t['east san gabriel'].loc[['pop_det_pov']])
     /int(crash_data19_t['san gabriel'].loc[['pop_det_pov']]))          #weight
    * int(crash_data19_t['east san gabriel'].loc[['pov_level']]))     # value
        ), decimals =2)

crash_data19_t
```
